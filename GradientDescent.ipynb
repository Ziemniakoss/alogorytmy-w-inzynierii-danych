# Gradient Descent

```julia
using Zygote
using LinearAlgebra
```

```julia
function GradientDescent(f, X; maxIterations = 1e5, α = 0.1, divideAlphaIfFIncreasesBy = 2.0, ϵ = 1e-7)
    fₙ = f(X)
    Xₙ = X
    for _ in 1:maxIterations
        ∇f = f'(Xₙ)
        if(norm(∇f) <= ϵ)
            break
        end
        Xₙ₊₁ = Xₙ - α * ∇f
        if(norm(Xₙ₊₁ - Xₙ) <= ϵ)
            break;
        end
        fₙ₊₁ = f(Xₙ₊₁)
        if fₙ₊₁ >= fₙ
            α /= divideAlphaIfFIncreasesBy
        end
        fₙ = fₙ₊₁
        Xₙ = Xₙ₊₁
    end
    
    return Xₙ;    
end
```

```julia
function f(x)
    x[1]^2
end
```

```julia
GradientDescent(f, [20])
@time GradientDescent(f, [20])
```
